```python
import matplotlib.pyplot as plt
```

```python
age,height = ([17,18,19,20,21], [130, 141, 151, 163, 174])


plt.figure(figsize=(10, 3))

plt.bar(age,height)

plt.show()
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)

plt.bar(age,height)

plt.subplot(1, 2, 2)

plt.barh(age,height)

plt.show()
```

```python
plt.figure(figsize=(10, 3))

plt.plot(age,height)

plt.show()
```
